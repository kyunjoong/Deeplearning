```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.array([2,4,6,8])
x2 = np.array([8,4,2,3])
y = np.array([81,93,91,97])
```

```python
# plot Data Distribution
fig  = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter3D(x1,x2,y);
plt.show()
```

```python
# initialize a1,a2 and b 
a1 = 0
a2 = 0 
b = 0
```

```python
# set learning rate and epoches 
lr = 0.014
epochs = 3001
```

Change to lr to epochs and find global minimum 

problems about gradient descent 
- local minimum (find local minimum not global minmum
- oscilation 
 if learning late small, it goes slow but It converges stably
 if learning late high, it goes fast but values are oscilate near converges 

```python
n =len(x1)
```

```python
for i in range(epochs):

  y_pred =  a1 * x1 +  a2 * x2 + b 
  error = y - y_pred

  a1_diff = (2/n) * sum(-x1 * (error))
  a2_diff = (2/n) * sum(-x2 * (error))
  b_diff = (2/n) * sum(-(error))

  a1 = a1 -lr * a1_diff
  a2 = a2 -lr * a2_diff
  b = b - lr * b_diff

  if i % 100 == 0:
    print("epoch = %.f, gradient_1 = %.04f, gradient_2 = %.04f, intercept = %.04f" %(i,a1,a2,b))
```
